```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
print(df.columns)
```

```python
plt.scatter(df.year, df.pci, color = "red", marker = "*")
plt.xlabel("year")
plt.ylabel("PCI")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df.pci)
```

```python
new_predict = reg.predict([[2020]])

print(new_predict)
```

```python
plt.scatter(df.year, df.pci, color = "red", marker = "*")
plt.xlabel("year")
plt.ylabel("PCI")
plt.plot(df.year, reg.predict(df[['year']]), color = "green")
```
